```python
alphanumeric_postal_codes = df[df['Postleitzahl'].str.contains(r'[a-zA-Z]', na=False)]
alphanumeric_postal_codes
```

```python
df['Postleitzahl'].fillna('', inplace=True)

# Apply the unary ~ operator after handling NaN values
df= df[~df['Postleitzahl'].astype(str).str.contains('A') & ~df['Postleitzahl'].astype(str).str.contains('Weidweg36')]
df
```

```python
#df['Postleitzahl'].replace('', 0, inplace=True)
df.loc[df['Postleitzahl'] == '', 'Postleitzahl'] = 0

df.loc[df['Postleitzahl'] == 0]
```

```python
df["Postleitzahl"] = df["Postleitzahl"].astype(float)
```

```python
df["Postleitzahl"] = df["Postleitzahl"].astype(int)
```

```python
df["Bruttoleistung der Einheit"] = df["Bruttoleistung der Einheit"].str.replace(',', '.')
df["Nettonennleistung der Einheit"] = df["Nettonennleistung der Einheit"].str.replace(',', '.')
df["Bruttoleistung der Einheit"] = df["Bruttoleistung der Einheit"].astype(float)
df["Nettonennleistung der Einheit"] = df["Nettonennleistung der Einheit"].astype(float)
```

```python
df["Gemeindeschlüssel"] = df["Gemeindeschlüssel"].fillna(0)
df["Gemeindeschlüssel"] = df["Gemeindeschlüssel"].astype(int)
```

```python
df["Betriebs-Status"] = df["Betriebs-Status"].astype(str)
```

```python
df["Koordinate: Breitengrad (WGS84)"] = df["Koordinate: Breitengrad (WGS84)"].fillna(0)
df["Koordinate: Breitengrad (WGS84)"] = df["Koordinate: Breitengrad (WGS84)"].str.replace(',', '.')
df["Koordinate: Breitengrad (WGS84)"] = df["Koordinate: Breitengrad (WGS84)"].astype(float)
```

```python
df["Koordinate: Längengrad (WGS84) )"] = df["Koordinate: Längengrad (WGS84)"].fillna(0)
df["Koordinate: Längengrad (WGS84)"] = df["Koordinate: Längengrad (WGS84)"].str.replace(',', '.')
df["Koordinate: Längengrad (WGS84)"] = df["Koordinate: Längengrad (WGS84)"].astype(float)

df["Nabenhöhe der Windenergieanlage"] = df["Nabenhöhe der Windenergieanlage"].fillna(0)
df["Nabenhöhe der Windenergieanlage"] = df["Nabenhöhe der Windenergieanlage"].str.replace(',', '.')
df["Nabenhöhe der Windenergieanlage"] = df["Nabenhöhe der Windenergieanlage"].astype(float)
df["Rotordurchmesser der Windenergieanlage"] = df["Rotordurchmesser der Windenergieanlage"].fillna(0)
df["Rotordurchmesser der Windenergieanlage"] = df["Rotordurchmesser der Windenergieanlage"].str.replace(',', '.')
df["Rotordurchmesser der Windenergieanlage"] = df["Rotordurchmesser der Windenergieanlage"].astype(float)
```

```python
df["Inbetriebnahmedatum der Einheit"] = df["Inbetriebnahmedatum der Einheit"].fillna(df["Registrierungsdatum der Einheit"])
```

```python
df["Inbetriebnahmedatum der Einheit"].isna().sum()
```

```python
df["Inbetriebnahmedatum der Einheit"]= pd.to_datetime(df['Inbetriebnahmedatum der Einheit'], format='mixed')
```

```python
df["Registrierungsdatum der Einheit"]= pd.to_datetime(df['Registrierungsdatum der Einheit'], format='mixed')
```

```python
df.drop(['Straße','Hausnummer','Gemarkung','Flurstück','Name des Windparks','Hersteller der Windenergieanlage','Typenbezeichnung','Hauptbrennstoff der Einheit','Nutzbare Speicherkapazität in kWh','Technologie der Stromerzeugung','Letzte Aktualisierung','Datum der endgültigen Stilllegung','Datum der geplanten Inbetriebnahme','\tMaStR-Nr. des Anlagenbetreibers','MaStR-Nr. der Genehmigung','Name des Anschluss-Netzbetreibers','MaStR-Nr. des Anschluss-Netzbetreibers','Netzbetreiberprüfung','Spannungsebene','MaStR-Nr. der Lokation','MaStR-Nr. der EEG-Anlage','EEG-Anlagenschlüssel','Inbetriebnahmedatum der EEG-Anlage','Installierte Leistung der EEG-Anlage','Zuschlagnummer (EEG/KWK-Ausschreibung)','MaStR-Nr. der KWK-Anlage','Inbetriebnahmedatum der KWK-Anlage','Elektrische KWK-Leistung','Thermische Nutzleistung in kW'], axis=1 , inplace=True)
```

## Upload to BigQuery

```python
#Replace column names

# List of new column names
new_column_names = [
    "market_registry_nb",
    "unit_name",
    "operational_status",
    "energy_source",
    "gross_power_bruttoleistung",
    "net_rated_power_nettonennleistung",
    "commissioning_date",
    "registry_date",
    "state",
    "zip_code",
    "town",
    "gemeindeschluessel",
    "coordinates_latitude",
    "coordinates_longitude",
    "nb_solar_modules",
    "orientation_solar_modules",
    "hub_height_wind",
    "rotor_diameter_wind",
    "unit_location",
    "owner_name_only_orgs",
    "full_or_partial_feed_in"
]

# Replace column names
df.columns = new_column_names
```

```python
from google.oauth2 import service_account
from google.cloud import bigquery
```

```python
# Define your BigQuery client
client = bigquery.Client()

# Set up BigQuery credentials
credentials = service_account.Credentials.from_service_account_file('/Users/juliankilchling/Downloads/energytransition-final-project-b07a66757bd2.json')

# Set up
project_id = 'energytransition-final-project'
dataset_id = 'energycharts'

table_id = f'{dataset_id}.market_registry'

#Pandas to_gbq function
df.to_gbq(destination_table=table_id, project_id=project_id, credentials=credentials, if_exists='replace')
```
